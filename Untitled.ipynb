```python
import pandas as pd

# Load the dataset
df = pd.read_csv("vehicles.csv")

# Check the shape of the dataset
print(df.shape)

# Display the first 5 rows
df.head()

# Get a summary of data types and missing values
df.info()

# Display percentage of missing values per column
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
print(missing_percent)

# List of columns that are clearly not useful
useless_cols = ['id', 'url', 'region_url', 'image_url', 'description', 'VIN', 'county', 'posting_date', 'size']

# Drop those columns from the dataframe
df = df.drop(columns=useless_cols)

# Fill categorical features with 'unknown'
for col in ['cylinders', 'condition', 'drive', 'paint_color', 'type']:
    df[col] = df[col].fillna('unknown')

# Fill important categorical features with unknown
for col in ['manufacturer', 'model', 'fuel', 'transmission', 'title_status']:
    df[col] = df[col].fillna('unknown')

# Fill numeric values with median
for col in ['year', 'odometer', 'lat', 'long']:
    df[col] = df[col].fillna(df[col].median())
```

```python
## EDA ##
import matplotlib.pyplot as plt
import seaborn as sns

# Check lowest prices
print(df['price'].value_counts().head(10))

# Check if there are extremely high prices
print(df['price'].max())

# Remove entries with price <= 100 or price > 100000
df = df[(df['price'] > 100) & (df['price'] < 100000)]

plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=100, kde=True)
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.xlim(0, 100000)
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x='year', y='price', data=df, alpha=0.3)
plt.title('Price vs Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.ylim(0, 100000)
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x='odometer', y='price', data=df, alpha=0.3)
plt.title('Price vs Odometer')
plt.xlabel('Mileage (odometer)')
plt.ylabel('Price')
plt.ylim(0, 100000)
plt.show
```

```python
import numpy as np

# Calculate average price per manufacturer
top_manufacturers = df['manufacturer'].value_counts().head(10).index
df_top_mf = df[df['manufacturer'].isin(top_manufacturers)]

plt.figure(figsize=(12, 6))
sns.barplot(x='manufacturer', y='price', data=df_top_mf, estimator=np.mean)
plt.title('Average Price by Top 10 Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.ylim(0, 60000)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='condition', y='price', data=df)
plt.title('Price by Vehicle Condition')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.ylim(0, 100000)
plt.show()
```
